### Question 01
Scientific Data Analysis with NumPy You are a scientist conducting research on climate change. You have collected temperature data from weather stations around the world stored in a NumPy array. Your task is to analyze the data using NumPy to:
1. Calculate the average temperature for each month across all stations.
2. Identify the station with the highest and lowest average temperature.
3. Determine the month with the largest temperature variation.
4. Visualize the temperature data using line plots or heatmaps.



<b>Use the following data set

```python
import numpy as np

# Sample temperature data for each month (in Celsius)
temperature_data = np.array([
    # January
    [10, 12, 15, 8, 9],
    # February
    [11, 13, 14, 9, 10],
    # March
    [12, 14, 16, 10, 11],
    # April
    [14, 16, 18, 12, 13],
    # May
    [16, 18, 20, 14, 15],
    # June
    [18, 20, 22, 16, 17],
    # July
    [20, 22, 24, 18, 19],
    # August
    [22, 24, 26, 20, 21],
    # September
    [20, 22, 24, 18, 19],
    # October
    [18, 20, 22, 16, 17],
    # November
    [15, 17, 19, 13, 14],
    # December
    [12, 14, 16, 10, 11]
])

# Station names for reference
station_names = ['Station A', 'Station B', 'Station C', 'Station D', 'Station E']
```

```python
print("Average temperature for each station",temperature_data.mean(axis=0))   # Mean temperature for each station
print("Average temperature for each month",temperature_data.mean(axis=1))   # Mean temperature for each month
```

```python
avg_tmp = temperature_data.mean(axis=0)
print("Station with highest average temperature is",station_names[np.argmax(avg_tmp)])
print("Station with lowest average temperature is",station_names[np.argmin(avg_tmp)])
```

```python
variance = temperature_data.var(axis=1)
print(f"Month no {variance.argmax()+1} has the largest temperature variation")
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.imshow(temperature_data.T, cmap='hot_r')
plt.colorbar()
plt.yticks(np.arange(len(station_names)), station_names)
plt.xticks(np.arange(12), np.arange(1, 13))
plt.xlabel('Month')
plt.ylabel('Station')
plt.title('Temperature Data')
plt.show()
```

```python
daily_returns = np.diff(stock_prices) / stock_prices[:-1]
print(daily_returns)
```

```python
mean_return = daily_returns.mean()
std_return = daily_returns.std()
var_return = daily_returns.var()
print(f"Mean: {mean_return:.7f}, Standard Deviation: {std_return:.7f}, Variance: {var_return:.7f}")
```

```python
print(f"Day with the highest return: {np.argmax(daily_returns) + 1}")
print(f"Day with the lowest return: {np.argmin(daily_returns) + 1}")
```

```python
cumulative_returns = np.cumprod(1 + daily_returns) - 1
print(cumulative_returns)
```

You are working on a project that involves image processing. In this assignment, you will utilize NumPy to implement various operations on an RGB image of your own choice. Follow the steps below:

- Select an Image:
- Choose an RGB image of your own choice. You can use any image. Ensure that the image is in RGB format.
- Convert RGB Image to Grayscale:
- Convert the grayscale image to a binary image using a threshold value.
- Apply Gaussian blur to the binary image to reduce noise.
- Detect edges in the blurred image using the Sobel operator.
- Compute the histogram of pixel intensities in the original grayscale image.


<br> Visualize the results of each operation to observe the effects on the image.
Submission Guidelines:

<br>Submit the selected RGB image in the submission as well.

```python
from PIL import Image
def load_image(image_path):
    image = Image.open(image_path)
    image_array = np.array(image)
    return image_array
image_array = load_image('jungle.jpg')
print(image_array.shape)
```

```python
plt.imshow(image_array)
plt.show()
```

```python
def rgb_to_gray(image):
    return np.dot(image[...,:3], [0.2, 0.5, 0.1])
gray_image = rgb_to_gray(image_array)
print(gray_image.shape)
plt.imshow(gray_image, cmap='gray')
plt.show()
```

```python
def binarize_image(image, threshold):
    return (image > threshold).astype(int)
threshold = 100
binary_image = binarize_image(gray_image, threshold)
print(binary_image)
plt.imshow(binary_image, cmap='gray')
plt.show()
```

```python
from scipy.ndimage import gaussian_filter
blurred_image = gaussian_filter(binary_image, sigma=1)
plt.imshow(blurred_image, cmap='gray')
plt.show()
```

```python
from scipy.ndimage import sobel
sobel_image = sobel(blurred_image)
plt.imshow(sobel_image, cmap='gray')
plt.show()
```

```python
histogram, bins = np.histogram(gray_image, bins=256, range=(0, 256))
plt.plot(histogram)
plt.title('Pixel Intensity Histogram')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.show()
```
